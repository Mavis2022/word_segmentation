```python
dfred=pd.read_csv('red_top1000.csv')
#åˆ—å‡ºæ€»å…±æœ‰å¤šå°‘ä¸ªä¸åŒå“ç‰Œï¼ˆunique values of the brand columnï¼‰
brands_total = dfred['brand'].unique()
print(brands_total)
```

```python
dfred.head()
```

```python
#åˆ é™¤contentæ˜¯ç©ºçš„é‚£äº›è¡Œ
dfred = dfred[dfred['content'].notna()]

###æ–‡æœ¬æ¸…æ´—ï¼ˆæ¸…æ´—åçš„columnåå­—ä¸ºclean_cmtï¼‰
import re
def process_text(text):
    text = re.sub(r"[0-9]+", '', text) 
    text = re.sub(r'\s+',' ',text)
    text = text.lower()
    text = re.sub(r'\d',' ',text)
    text = re.sub(r'\s+',' ',text)
    text = ''.join(e for e in text if e.isalnum())
    return text

content = dfred['content'].dropna() #æŠŠç©ºçš„contentå»æ‰
processed_content = []
for i in range(len(content)):
    processed_content.append([process_text(content.iloc[i])])
```

```python
import numpy as np
dfred["clean_cmt"] = np.array(processed_content)
```

```python
###åŠ å…¥stopwords
ss = pd.read_excel('redè¯é¢‘ç»Ÿè®¡_æ ‡æ³¨.xlsx',engine = 'openpyxl')
stopwords = ss[ss['Unnamed: 2'] == 1]
stopwordlist = list(stopwords['Unnamed: 0'])
stopwordlist.extend(['é£å»','ç¬‘å“­','å·ç¬‘','å“­æƒ¹','æ´¾å¯¹','å¾®ç¬‘','å®³ç¾'])

#åˆ›å»ºcontent dataframeï¼Œåˆ—åç§°ï¼štextï¼Œå†…å®¹ï¼šæ–‡æœ¬å¤„ç†åçš„content
content = pd.DataFrame(processed_content,columns = ['text'])

adj = []
allow_pos = ('na') #é€‰å‡ºå½¢å®¹è¯&åè¯
for i in range(len(content)):
    lines = content.iloc[i].values.tolist()
    content1 = "".join(lines)
    words = jieba.analyse.extract_tags(content1, topK=50, withWeight=False, allowPOS=(allow_pos))
    for w in words:
        if w in stopwordlist:
            words.remove(w)
    adj.append([i,",".join(words)])
```

```python
#åœ¨åŸdataframeçš„åŸºç¡€ä¸Šæ›´æ–°dataframeï¼šåŠ å…¥ä¸€åˆ—åŒ…å«æ‰€æœ‰å½¢å®¹è¯&åè¯
dfred['adj'] = adj
```

```python
###æ­¤å¤„ä»¥ç¬¬ä¸€ä¸ªå“ç‰Œç¤ºä¾‹
brand0 = []
for index,row in dfred.iterrows():
    if brands_total[0] in str(row['brands']):  #å°†æ‰€æœ‰æ˜¯æ­¤å“ç‰Œçš„rowæ‘˜å‡º
        brand0.append(row['adj']) #å°†å“ç‰Œå¯¹åº”çš„åˆ‡è¯ç»“æœæ‘˜å‡º
        
text = ",".join(i for i in adj)
# æ³¨æ„ï¼šæ”¹æˆè‡ªå·±çš„å­—ä½“è·¯å¾„ï¼ï¼ï¼ï¼ï¼
font = r'AaLingGanHei45J-2.ttf'   # æ”¹æˆç›¸åº”çš„å­—ä½“çš„è·¯å¾„å³å¯

wc1600by1000 = WordCloud(collocations=False, background_color='white',font_path=font, max_words=2000,width=1600, height=1000, margin=4).generate(text.lower())
wc800by500 = WordCloud(collocations=False, background_color='white',font_path=font, max_words=2000,width=800, height=500, margin=4).generate(text.lower())

plt.imshow(wc1600by1000)
plt.imshow(wc800by500)
plt.axis("off")
plt.show()
#wc.to_file('redcmt.png')  # æŠŠè¯äº‘ä¿å­˜ä¸‹æ¥
```
