```python
dianp = []
for i in range(len(dfxq)):
    if 'negative' in dfxq['ratings'][i]:
        s = 0
    else:
        s = 1
    dianp.append(s)
dfxq['dianp'] = dianp
```

```python
human_dianp = []
for i in range(len(dfresult_xq)):
    if dfresult_xq['score'][i] <= 3:
        a = 0
    else:
        a = 1
    human_dianp.append(a)

dfresult_xq['human_dianp'] = human_dianp
```

```python
import re
s = list(int(s) for s in re.findall(r'-?\d+\.?\d*', dfxq['ratings'][0][12:19]))
```

```python
def convert(list):
      
    # Converting integer list to string list
    s = [str(i) for i in list]
      
    # Join list items using join()
    res = int("".join(s))
      
    return(res)
```

```python
clean_dg = []
for i in range(len(dfxq.head(110))):
    s = list(int(s) for s in re.findall(r'-?\d+\.?\d*', dfxq['ratings'][i][12:19]))
    s = convert(s)
    clean_dg.append(s)
    
```

```python
clean = []
for i in range(len(dfxq)):
    s = list(int(s) for s in re.findall(r'-?\d+\.?\d*', dfxq['ratings'][i][12:19]))
    s = convert(s)
    if s >= 4:
        result = 'positive'
    elif s == 3:
        result = 'neutral'
    else:
        result = 'negative'
    clean.append(result)
```

```python
clean_model_bi = []
for i in range(len(dfresult_xq)):
    s = int(dfresult_xq['model_star'][i])
    if s >= 3:
        result = 1
    else:
        result = 0
    clean_model_bi.append(result)
dfresult_xq['model_bi'] = clean_model_bi
```

```python
clean_human_bi = []
for i in range(len(dfresult_xq)):
    s = int(dfresult_xq['score'][i])
    if s >= 3:
        result = 1
    else:
        result = 0
    clean_human_bi.append(result)
dfresult_xq['human_bi'] = clean_human_bi
```
