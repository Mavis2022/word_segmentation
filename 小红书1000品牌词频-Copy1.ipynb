## 小红书词云代码

```python
dfred=pd.read_csv('red_top1000.csv')
#列出总共有多少个不同品牌（unique values of the brand column）
brands_total = dfred['brand'].unique()
print(brands_total)
```

```python
dfred.head()
```

```python
#删除content是空的那些行
dfred = dfred[dfred['content'].notna()]

###文本清洗（清洗后的column名字为clean_cmt）
import re
def process_text(text):
    text = re.sub(r"[0-9]+", '', text) 
    text = re.sub(r'\s+',' ',text)
    text = text.lower()
    text = re.sub(r'\d',' ',text)
    text = re.sub(r'\s+',' ',text)
    text = ''.join(e for e in text if e.isalnum())
    return text

content = dfred['content'].dropna() #把空的content去掉
processed_content = []
for i in range(len(content)):
    processed_content.append([process_text(content.iloc[i])])
```

```python
import numpy as np
dfred["clean_cmt"] = np.array(processed_content)
```

```python
###加入stopwords
ss = pd.read_excel('red词频统计_标注.xlsx',engine = 'openpyxl')
stopwords = ss[ss['Unnamed: 2'] == 1]
stopwordlist = list(stopwords['Unnamed: 0'])
stopwordlist.extend(['飞吻','笑哭','偷笑','哭惹','派对','微笑','害羞']) #加入额外的自定义stopwords
```

```python
#创建content dataframe，列名称：text，内容：文本处理后的content
content = pd.DataFrame(processed_content,columns = ['text'])
```

```python
allow_pos = ('na') #选出形容词&名词（此处只以品牌1为例）
brand0 = [] ##brand0储存品牌1对应的所有文本（清理后）
jieba0 = [] ##brand0储存品牌1对应的所有文本的切词结果（清理后）
for index,row in dfred.iterrows():
    if brands_total[0] in str(row['brand']):  #将所有是品牌1的row摘出
        brand0.append(row['clean_cmt']) #将品牌1对应的品论文本摘出
```

```python
brand0
```

```python
for i in range(len(brand0)): #loop品牌1的每一条评论
    lines = brand0[i].split()
    content1 = "".join(lines)
    words = jieba.analyse.extract_tags(content1, topK=50, withWeight=False, allowPOS=(allow_pos))
    for w in words:
        if w in stopwordlist:
            words.remove(w)
    jieba0.append([i,",".join(words)])
```

```python
jieba0
```

```python

dfbrand0 = dfred.head(len(brand0)) #dfbrand0：为第一个品牌单独创立一个dataframe
dfjieba0 = pd.DataFrame(jieba0) #将第一个品牌的分词结果创一个dataframe
jieba0 = dfjieba0[1] #更新jieba0为仅有关键词，无index（用dfjieba0中关键词那一列）
dfbrand0['jieba'] = jieba0 #在原dfbrand0的基础上更新dataframe：加入一列包含所有形容词&名词
dfbrand0
```
